```python
import re 

#================================================================================================#
#  ============================================================================================  #
#                               THEORY OF AUTOMATA SEMSTER PROJECT                               #
#                                                                                                #
#                                  REGULAR EXPRESSION ENGINE                                     #
#                                                                                                #
#               CODE AUTHORs: ARVIND KUMAR(SP17-BSCS-0042) & MUDASIR ALI(SP17-BSCS-0082)         #
#  ============================================================================================  #
#================================================================================================#

filename ="E:\\4TH SEMESTER\TOA\RegexText.txt"   #OPEN THE TEXT FILE 
file = open(filename,'r')
lines=file.readlines()                           #READ THE TEXT FILE LINE BY LINE 
count=0
i=0
print('enter what you want to search from text file :')
x = input()

#===============================for emails addresses======================================
if x=="emails":
  for line in lines:
      i=i+1
      emails = re.findall(r"[a-zA-Z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z\.]{2,}", line)
   
      if emails:
         count=count+1
         print(emails,i)
         print(count)  #COUNT THE NUMBER OF EMAILS
     # print(i)  
      

#==================================for date ================================================
elif x=="date":

    for line in lines:
         i=i+1
        # MATCH THE DATE PATTREN IN EACH LINE
         date= re.findall(r"\d{2,4}[.-/]\d{1,2}[.-/]\d{2,4}",line)
         if date: 
             count=count+1
             print(date,i)
             print(count)   #COUNT THE NUMBER OF DATE
          
    #print(line.strip())
    
    
 #=================================FOR URLS==================================================
elif x=="urls":
    for line in lines:
         i=i+1
        # MATCH THE  URLS PATTREN IN EACH LINE
         urls= re.findall(r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-zA-Z][0-9a-zA-Z]))+",line)
            
         if urls: 
            count=count+1
            print(urls,i)
            print(count)    #COUNT THE NUMBER OF URLS 
    #print(line.strip())

#==============================contactNO======================================================    

elif x=="contactNo":
    for line in lines:
         i=i+1
        
        # MATCH THE CONTACT NUMBER PATTREN IN EACH LINE
         contactNo= re.findall(r"(1[-]\d\d\d[-]\d\d\d[-]\d\d\d\d|\d\d\d[-]\d\d\d\d)",line)
                       
         if contactNo: 
            count=count+1    
            print(contactNo,i)
            print(count)      #COUNT THE NUMBER OF CONTACT NUMBER  
    #print(line.strip())
```
